```python
import pandas as pd 
import polars as pl
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import des données
ift = pd.read_csv(
    "./../../data/raw/fre-324510908-adonis-ift-2022-v04112024.csv",
    sep=";", low_memory=False, index_col=0
)
print(ift.shape)
ift.head()
```

```python
# Types de cultures
ift["cod_c_maj"].unique()
```

```python
# Toutes les cultures
liste_cultures = set(ift["c_maj"]) | set(ift["c_ift_hbc"]) | set(ift["c_ift_h"])
print("Nombre de cultures différentes :", len(liste_cultures))
print("Cultures différentes :", liste_cultures)
```

```python
ift.columns
```

```python
# Liste des culture avec codes cultures
liste_cultures_df = pd.DataFrame(
    {
        "culture": list(liste_cultures),
        "code_culture": [None] * len(liste_cultures)
    }
)   

for c in liste_cultures:
    if c in ift["c_maj"].unique():
        code = ift[ift["c_maj"] == c]["cod_c_maj"].unique()[0]
        liste_cultures_df.loc[liste_cultures_df["culture"] == c, "code_culture"] = code
        print(f"{c} : {code}")
    elif c in ift["c_ift_hbc"].unique():
        code = ift[ift["c_ift_hbc"] == c]["cod_c_hbc"].unique()[0]
        liste_cultures_df.loc[liste_cultures_df["culture"] == c, "code_culture"] = code
        print(f"{c} : {code}")
    elif c in ift["c_ift_h"].unique():
        code = ift[ift["c_ift_h"] == c]["cod_c_h"].unique()[0]
        liste_cultures_df.loc[liste_cultures_df["culture"] == c, "code_culture"] = code 
        print(f"{c} : {code}")         
```
